```python
kv=np.array([[1,0,-1]*3]).reshape((3,3))
```

```python
kh=(kv.copy()).transpose()
```

1.NO PADDING:
stride=1

```python
rm=np.zeros((l-2,b-2))
```

i)Vertical line filter

```python
fl=[]
for k in [kv,kh]:
  rm=np.zeros((l-2,b-2))
  for i in range(l-2):
    for j in range(b-2):
      w=np.array((list(img[i,j:j+3])+list(img[i+1,j:j+3])+list(img[i+2,j:j+3]))).reshape((3,3))
      rm[i,j]=(w*k).sum()
  fl.append(rm)
```

```python
cv2_imshow((fl[0]+fl[1])/2)
```
